# `**# TO FIND THE EQUATION OF LINE ::WHEN GIVEN TWO POINTS::**`


```
import matplotlib.pyplot as plt
import numpy as np
def find_line(x1,y1,x2,y2):
  #calculate the slope (m) and intercept (c)
  m=(y2-y1)/(x2-x1)
  c=y1-m*x1
  return m,c

def plot_line(x1,y1,x2,y2):
  #calculate line
  m,c=find_line(3,5,9,10)
  x1==3,y1==5,x2==9,y2==10
  x=np.linspace(x1-10,x2+10,400)
  y=m*x+c
  plt.plot(x,y,label=f"y={m}x+{c}")
  plt.scatter([x1,x2],[y1,y2],label="given points")
  print(find_line(3,5,9,10))
  print("the equation of the line")
find_line(3,5,9,10)
plot_line(3,5,9,10)
```

# **# TO FIND THE EQUATION FO LINE:: WHEN GIVEN FIVE POINTS::**:



```
import numpy as np #numerical python #n-Dim array
import matplotlib.pyplot as plt #plotting
```

```
x = [1,2,3,4,5]
y = [2,4,5,4,5]
```

```
#mean_x , mean_y
n= len(x)
mean_x = np.mean(x)
mean_y = np.mean(y)
print(mean_x)
print(mean_y)
#print(n)
```

```
nume = 0
deno = 0
for i in range(n):
  nume += (x[i] - mean_x)*(y[i] - mean_y)
  deno += (x[i] - mean_x)**2
m = nume/deno
c = mean_y - (m*mean_x)
print(m)
print(c)
```

```
plt.scatter(x,y,label='scatter plot')
plt.plot(x,m*np.array(x)+c,label='line')
plt.legend()
plt.show()
```

```
max_x = np.max(x) + 10
min_x = np.min(y) - 10

a= np.linspace(min_x,max_x,100)
#print(a)
b= m*a + c
#print(b)
plt.scatter(x,y, color = 'red')
plt.plot(a,b,color = 'blue')
plt.show()
```
